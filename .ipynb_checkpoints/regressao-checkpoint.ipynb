### Cria conjunto de teste e conjunto de treino pelo PANDAS

```python
#conjunto com o pandas
dataset = datapd
K = 20 
N = np.size(dataset) #amostras
M = N-K #novo dataset para testes
X = np.empty((M,K))
y = np.empty((M,1))

for i in range(M):
    for k in range(K):
        X[i][k] = dataset.values.flatten()[(i+K) - (k + 1)]
    y[i][0] = (dataset.values.flatten()[i+K])
```

```python
amostra_teste  = 5*12 # define o conjunto de testes

X_treino = np.array(X[:-amostra_teste])
y_treino = np.array(y[:-amostra_teste])

X_teste = np.array(X[-amostra_teste:])
y_teste = np.array(y[-amostra_teste:])
```

```python
Phi_treino = np.column_stack((np.ones(X_treino.shape[0]), X_treino))
Phi_teste = np.column_stack((np.ones(X_teste.shape[0]), X_teste))
```

```python
Phi_treino.shape, y_treino.shape, Phi_teste.shape, y_teste.shape
```

### RMSE critério de raiz do erro médio quadratico

```python
def rmse(saida, meta):
    error = np.subtract(saida, meta)
    error = np.square(error)
    error = np.sum(error)/np.size(error)
    return np.sqrt(error)
```

### Regressão linear

```python
reg = lambda W, Phi: np.dot(Phi, W)
```

## com MMQ

```python
mmq = lambda Phi, y: np.dot(np.dot(np.linalg.inv(np.dot(Phi.T, Phi)), Phi.T), y)
```

```python
w = mmq(Phi_treino, y_treino)
```

```python
saida_treino = reg(w, Phi_treino)
saida_teste = reg(w, Phi_teste)
```

```python
print("RMSE do conjunto de treino: {}\nRMSE do conjunto de teste: {}".format(rmse(saida_treino, y_treino), rmse(saida_teste, y_teste)))
```

```python
plt.plot(saida_teste)
plt.plot(y_teste, '-r')
plt.xlabel('Mês')
plt.legend(('saida', 'objetivo'))
plt.title('Comparação entre saída e rotulos no conjunto de teste.')
plt.savefig('saidaerotulos.png', bbox_inches='tight')
plt.show()
```

##  wrapper, backward elimination, validação cruzada, k-fold e regularização L2

```python
def get_i(Phi):
    N = Phi.shape[1]
    ret = np.eye(N)
    ret[0,0] = 0
    return ret
```

```python
def mmq_l2(Phi, y, l):
    Iprime = get_i(Phi)
    ret = np.dot(np.dot(np.linalg.inv(np.dot(Phi.T, Phi) + l*Iprime), Phi.T), y)
    return ret
```

### K-fold rmse e lambda usando

```python
def kfold(Phi, y, L, lamb_vec, v=0):
    N = Phi.shape[0]
    min_rmse = 9999
    min_lamb = None
    M = N//L # num de pastas
    for l in lamb_vec:
        error = 0
        for i in range(L):
            train_slc = list(range(i*M)) + list(range((i+1)*M, N))
            test_slc = list(range(i*M,(i+1)*M))
            w = mmq_l2(Phi[train_slc], y[train_slc], l)
            out = reg(w, Phi[test_slc])
            error += rmse(out, y[test_slc])
            if v:
                print(error)
        error /= L
        if v:
            print('[', error, ']')
        if error < min_rmse:
            min_rmse = error
            min_lamb = l
            
    return (min_rmse, min_lamb)
```

### Wrapper - backward elimination

```python
def wrapper(Phi, y, K, L, lamb_vec, v=1):
    idx_vec = list(range(21))
    idx_mask = [1]*21
    mask = [e for e, f in zip(idx_vec, idx_mask) if f]
    ret = []
    (error, lamb) = kfold(Phi, y, L, lamb_vec, v>1)
    ret.append([0, error, lamb, mask])
    if v > 0:
        print("menor RMSE = {:13.10f} | lambda = {} | Elementos: ".format(error, lamb), end="")
        print(mask)
        print()
    for i in range(1, K):
        min_error = 9999
        for j in mask[1:]:
            idx_mask[j] = 0
            mask = [e for e, f in zip(idx_vec, idx_mask) if f]
            (error, lamb) = kfold(Phi[:,mask], y, L, lamb_vec, v>1)
            idx_mask[j] = 1
            if error < min_error:
                min_error = error
                min_l = lamb
                min_idx = j
            if v > 1:
                print("\tN = {} | menor RMSE = {:.4f} | drop= {} | lambda = {}".format(i, error, j, lamb))
        idx_mask[min_idx] = 0
        mask = [e for e, f in zip(idx_vec, idx_mask) if f]
        ret.append([i, min_error, min_l, mask])
        if v > 0:
            print("menor RMSE = {:13.10f} | lambda = {} | Elementos: ".format(min_error, min_l), end="")
            print(mask)
            print()
    return ret
```

```python
lamb_vec = [0, 10, 30, 60, 100, 300, 600, 1000, 3000, 6000, 10000, 30000, 60000]
vec = wrapper(Phi_treino, y_treino, 20, 5, lamb_vec, 1)
```

```python
vec = np.array(vec)
idx = np.argmin(vec[:,1])
print(vec[idx]) # pega melhor rmse
```

```python
plt.plot(vec[:,0], vec[:,1])
plt.plot(vec[idx,0], vec[idx,1], 'rx')
plt.xlabel("Número de atributos retirados")
plt.ylabel("RMSE mínimo")
plt.title("RMSE mínimo para número de atributos retirados")
plt.savefig('backward.png', bbox_inches='tight')
plt.show()
```

```python
w = mmq_l2(Phi_treino[:,vec[idx,3]], y_treino, vec[idx, 2])
```

```python
saida_treino = reg(w, Phi_treino[:,vec[idx,3]])
saida_teste_wrapper = reg(w, Phi_teste[:,vec[idx,3]])
```

```python
print("RMSE do conjunto de treino: {}\nRMSE do conjunto de teste: {}".format(rmse(saida_treino, y_treino), rmse(saida_teste_wrapper, y_teste)))
```

```python
plt.plot(saida_teste_wrapper)
plt.plot(y_teste, '-r')
plt.xlabel('Mês')
plt.legend(('saida', 'objetivo'))
plt.title('Comparação entre saída e rotulos no conjunto de teste.')
plt.savefig('wrapper.png', bbox_inches='tight')
plt.show()
```
