## Aplicando a MLP

```python
classes = 1
dimen_entrada = 2 
epocas = 100
passada = 10

for H in [3, 6, 10, 30, 60, 100]:
    print("H = ", H)
    treino_losses = []
    val_losses = []
    
    model = torch.nn.Sequential(
        torch.nn.Linear(dimen_entrada, H),
        torch.nn.ReLU(),
        torch.nn.Linear(H, classes),
        torch.nn.Sigmoid(),
    )
    
    optimizer = optim.Adam(model.parameters())
    
    for epoca in range(epocas):
        model.train()
        for idx, (data, target) in enumerate(zip(torch.tensor(X_train, dtype=torch.float), torch.tensor(y_train, dtype=torch.float))):
            optimizer.zero_grad()
            output = model(data)
            loss = F.binary_cross_entropy(output, target)
            loss.backward()
            optimizer.step()
        
        if epoca % passada == 0:
            print('Epoca de treinamento: {}, perda: {:.6f}'.format(epoca, loss.item()))
        treino_losses.append(loss.item())
        
        model.eval()
        val_loss = 0
        acerto = 0
        with torch.no_grad():
            for data, target in zip(torch.tensor(X_val, dtype=torch.float), torch.tensor(y_val, dtype=torch.float)):
                output = model(data)
                val_loss += F.binary_cross_entropy(output, target, size_average=False).item()
                pred = torch.round(output)
                acerto += (1 if pred == target else 0)
                
        val_loss /= len(y_val)
        val_losses.append(val_loss)
        
    print('\nDados de validação: média de perdas: {:.4f}, acurácia: {}/{} ({:.0f}%)\n'.format(val_loss, acerto, len(y_val), 100. * acerto / len(y_val)))
    
    plt.plot(treino_losses, color='blue')
    plt.plot(val_losses, color='red')
    plt.legend(["treino", "validação"])
    plt.yscale("log")
    plt.title('Curva de aprendizado com {} neuronios na camada do meio'.format(H))
    plt.xlabel('Exemplos de treinamento vistos')
    plt.ylabel('Entropia cruzada loss')
    plt.show()
```
